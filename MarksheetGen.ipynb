```python
print("\n\t\t <<< Welcome To Sir Syed University of Engineering & Technology >>>\n")
login = False
admin = ['admin','admin123']
logout = False

subjects = []
stu1 = {'Name':"Easha", 'Father name':'Asif','Rollnum':'89', 'Semester':'3'}
stu2 = {'Name':"Dua", 'Father name':'Faisal','Rollnum':'103', 'Semester':'3'}
stu3 = {'Name':"Taha", 'Father name':'Qaiser','Rollnum':'120', 'Semester':'1'}
stu4 = {'Name':"Hannan", 'Father name':'Shahid','Rollnum':'92', 'Semester':'5'}
stu5 = {'Name':"Furqan", 'Father name':'Ayaz','Rollnum':'105', 'Semester':'7'}
stuInfo = {"student1": stu1,"student2": stu2,"student3": stu3,"student4": stu4,"student5": stu5}
count = MTotal = 0
midmarks = []
semmarks = []
def data_entry(name,fname,rnum,sem):
    newStudent = input("Enter Key of Student: ")
    new_data = {'Name':name, 'Father name':fname, 'Rollnum':rnum, 'Semester':sem}
    stuInfo[newStudent] = new_data
    print("\n\t\t**** Your data has been entered successfully ****\n")

while not login:
    user = input("Enter Your Username: ")
    pswrd = input("Enter Password: ")

    if user == admin[0] and pswrd == admin[1]:
        while not logout:

            print("\n\t <<< Welcome Admin >>>\n")
            login = True
            print("1: New Data Entry")
            print("2: Update Stored Marks")
            print("3: Display Marksheet")
            print("4: Print Stored Data ")
            print("5: Logout The Admin")
            print("\n\t\t-------------------------------\n")
            choice = int(input("Enter your choice: "))

            if choice >= 1 and choice <= 5:
                if choice == 1:
                    print("\n\t\t**** INFORMATION OF STUDENT ****\n")
                    name = input("Enter name: ")
                    fname = input("Enter father's name: ")
                    rnum = int(input("Enter roll number: "))
                    sem = int(input("Enter semester: "))
                    data_entry(name, fname, rnum, sem)
                elif choice == 2:
                    print("\n\t\t**** UPDATE STUDENT INFORMATION ****\n")

                    for stu_k, stu_v in stuInfo.items():
                        print("\t\t <<<<", stu_k, ">>>\n")
                        for i in stu_v:
                            print(i, ":", stu_v[i], end=" \n")

                    newname = input("Enter key of Student You Want To Update: ")
                    for stu_k, stu_v in stuInfo.items():
                        if stu_k == newname:
                            print("\t\t <<<<", stu_k, ">>>\n")
                            for i in stu_v:
                                print(i, ":", stu_v[i], end=" \n")

                    print("\n\t\t <<<< PLEASE ENTER UPDATED INFORMATION >>>>")
                    newval = input("Enter Updated Name: ")
                    newFa = input("Enter Father's Name: ")
                    newRnum = input("Enter New Student Roll No: ")
                    newSem = input("Enter Updated Semester: ")
                    stuInfo[newname] = {"Name":newval, "Father name":newFa,"Rollnum": newRnum,"Semester":newSem}

                    for stu_k, stu_v in stuInfo.items():
                        print("\t\t <<<<", stu_k, ">>>\n")
                        for i in stu_v:
                            print(i, ":", stu_v[i], end=" \n")

                elif choice == 3:
                    print("\n\t\t**** GENERATE STUDENT MARKSHEET ****\n")
                    name = input("Enter name: ")
                    fname = input("Enter father's name: ")
                    rnum = int(input("Enter roll number: "))
                    sem = int(input("Enter semester: "))
                    sub = int(input("How Many Subject's Student Enrolled in: "))

                    if sem > 0 and sem <= 8:
                        for i in range(sub):
                            subj = input("Enter Subject Name: ")
                            subjects.append(subj)

                        for subject in subjects:
                            Midmark = int(input("Enter MidTerm Marks of subject " + subject + " out of 30: "))
                            Semmmark = int(input("Enter Semester Marks of subject " + subject + " out of 70: "))
                            midmarks.append(Midmark)
                            semmarks.append(Semmmark)
                        print("\n\n\t\t********************** YOUR RESULT *********************\n\n")
                        print("\t\tSEMESTER: ", sem)
                        print("\n\t\tNAME:        ", name, "\t\tFATHER NAME: ", fname)
                        print("\n\t\tROLL NUMBER: ", rnum)

                        print("\n\t\t SUBJECTS           MARKS1     MARKS2    TOTAL")
                        for (x, y, z) in zip(subjects, midmarks, semmarks):
                            print("\n\t\t", x, "             ", y, "       ", z, "       ", y + z)
                            MTotal += y + z
                            if y + z < 44:
                                count = count + 1

                        if count == 0:
                            print("\n\t\tTOTAL MARKS: ", MTotal, "\t\tRESULT: PASS")
                        else:
                            print("\n\t\tTOTAL MARKS:", MTotal, "\t\tRESULT: FAIL")

                    else:
                        print("\n\t\t**** PLEASE ENTER VALID SEMESTER ****")
                elif choice == 4:
                    print("\n\t\t**** STUDENT General Information ****\n")
                    for stu_k,stu_v in stuInfo.items():
                        print("\t\t <<<<",stu_k,">>>\n")
                        for i in stu_v:
                            print(i,":",stu_v[i],end=" \n")

                elif choice == 5:
                    sure = input("Are you sure you want to logout?(y/n)")
                    if sure == 'Y' or sure == 'y':
                        print("You're Successfully Logout!!!")
                        logout = True
                        break
                    elif sure == 'N' or sure == 'n':
                        logout = False
                    else:
                        print("Error!")
                        logout = True

            else:
                print("\n\t\t**** PLEASE ENTER VALID CHOICE ****")
                continue

    else:
        print("**** INCORRECT USERNAME OR PASSWROD PLEASE TRY AGAIN ****\n")
        continue
```
